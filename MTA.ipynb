### Standardizing Reporting Categories for Analysis

```python
# Define mapping rules
mapping = {
    "Crew Availability": "Other",
    "Police & Medical": "Persons on Trackbed/Police/Medical",
    "Infrastructure & Equipment": "Signals",
    "Operating Conditions": "Stations and Structure",
    "Planned ROW Work": "Subway Car",
    "External Factors": "Track"
}
# Create a new column (incident_category) to store the mapped incident categories
delay["incident_category"] = delay["reporting_category"].map(mapping)


delay.head()
```

## Data Analysis

### Reasons for Delays

```python
# General reasons for delays
general_delay_reasons = delay.groupby('reporting_category')['delays'].sum().sort_values(ascending=False)
general_delay_reasons = general_delay_reasons.astype(float).apply(lambda x: "{:,.0f}".format(x))
print("General reasons for delays:\n", general_delay_reasons)
```

```python
# Delays by division and line
division_line_delay = delay.groupby(['division', 'line'])['delays'].sum().sort_values(ascending=False)
division_line_delay = division_line_delay.astype(float).apply(lambda x: "{:,.0f}".format(x))
print("\nDelays by division and line:\n", division_line_delay)
```

### Common Major Incident Between Pre and Post-Pandemic



```python
# Convert the 'count' column to numeric type
incident_pre['count'] = pd.to_numeric(incident_pre['count'], errors='coerce')
incident_post['count'] = pd.to_numeric(incident_post['count'], errors='coerce')

# Group the incident_pre dataframe by 'category' and calculate the total 'count' for each category.
pre_summary = incident_pre.groupby('category')['count'].sum().reset_index()
pre_summary = pre_summary.sort_values(by='count', ascending=False)


post_summary = incident_post.groupby('category')['count'].sum().reset_index()
post_summary = post_summary.sort_values(by='count', ascending=False)


print("Pre-pandemic incident totals:")
print(pre_summary)

print("\nPost-pandemic incident totals:")
print(post_summary)
```

### Common Reasons for Delays on Weekdays and Weekends


```python
# Comparision Delays on Weekdays and Weekends for incident_pre
comparison_pre = (
    incident_pre.groupby(['category', 'day_type'])['count'].sum()
    .unstack(fill_value=0)
    .reset_index()
)

# Rename columns for better understanding
comparison_pre.columns = ['incident_category', 'delays_weekday', 'delays_weekend']

# Sort the comparison dataframe by weekday delays
comparison_pre = comparison_pre.sort_values(by='delays_weekday', ascending=False)

# Display the comparison dataframe
print("Comparison for Pre-Pandemic Incidents:")
print(comparison_pre)
```

```python
# Comparision Delays on Weekdays and Weekends for incident_post
comparison_post = (
    incident_post.groupby(['category', 'day_type'])['count']
    .sum()
    .unstack(fill_value=0)
    .reset_index()
)

# Rename columns for better understanding
comparison_post.columns = ['incident_category', 'delays_weekday', 'delays_weekend']

# Sort the comparison dataframe by weekday delays
comparison_post = comparison_post.sort_values(by='delays_weekday', ascending=False)

# Display the comparison dataframe
print("Comparison for Post-Pandemic Incidents:")
print(comparison_post)
```

### Comparision Delays on Weekdays and Weekends by Division and Line

```python
# Comparision Delays on Weekdays and Weekends for incident_pre (grouped by division and line)
comparison_pre = (
    incident_pre.groupby(['division', 'line', 'day_type'])['count']
    .sum()
    .unstack(fill_value=0)
    .reset_index()
)

# Rename columns for better understanding
comparison_pre.columns = ['division', 'line', 'delays_weekday', 'delays_weekend']

# Sort the comparison dataframe by weekday delays
comparison_pre = comparison_pre.sort_values(by='delays_weekday', ascending=False)

# Display the comparison dataframe
print("Comparison for Pre-Pandemic Incidents (Division & Line):")
print(comparison_pre)
```

```python
# Comparision Delays on Weekdays and Weekends for incident_post (grouped by division and line)
comparison_post = (
    incident_post.groupby(['division', 'line', 'day_type'])['count']
    .sum()
    .unstack(fill_value=0)
    .reset_index()
)

# Rename columns for better understanding
comparison_post.columns = ['division', 'line', 'delays_weekday', 'delays_weekend']

# Sort the comparison dataframe by weekday delays
comparison_post = comparison_post.sort_values(by='delays_weekday', ascending=False)

# Display the comparison dataframe
print("Comparison for Post-Pandemic Incidents (Division & Line):")
print(comparison_post)
```

### Monthly Analysis of Incident Trends (Pre-Pandemic)


```python
# Date processing
incident_pre['date'] = pd.to_datetime(incident_pre['date'])
incident_pre['month_num'] = incident_pre['date'].dt.month
incident_pre['month_name'] = incident_pre['date'].dt.month_name()


# Group data by month
monthly_trends = incident_pre.groupby('month_name')['count'].sum().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])


# Plot Monthly Trends
plt.figure(figsize=(10, 6))
monthly_trends.plot(kind='bar', color='skyblue', title='Monthly Trends in Incidents (Pre-Pandemic)')
plt.xlabel('Month')
plt.ylabel('Total Incidents')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


```

### Monthly Analysis of Incident Trends (Post-Pandemic)

```python
# Date processing
incident_post['date'] = pd.to_datetime(incident_post['date'])
incident_post['month_num'] = incident_post['date'].dt.month
incident_post['month_name'] = incident_post['date'].dt.month_name()


# Group data by month
monthly_trends = incident_post.groupby('month_name')['count'].sum().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])


# Plot Monthly Trends
plt.figure(figsize=(10, 6))
monthly_trends.plot(kind='bar', color='skyblue', title='Monthly Trends in Incidents (Post-Pandemic)')
plt.xlabel('Month')
plt.ylabel('Total Incidents')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Analyzing Subway Incidents Post-Pandemic: Clustering by Frequency

```python
# Feature selection: total number of incidents per line
features = incident_post.groupby('line')['count'].sum().reset_index()
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features[['count']])

# K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
features['cluster'] = kmeans.fit_predict(scaled_features)

# Visualization
plt.figure(figsize=(12, 8))
plt.scatter(features['line'], features['count'], c=features['cluster'], cmap='viridis')

# Adjust X-axis label rotation and spacing
plt.xticks(rotation=45, fontsize=10)

plt.title('Clustering of Subway Lines by Incident Frequency', fontsize=14)
plt.xlabel('Line', fontsize=12)
plt.ylabel('Incident Count', fontsize=12)
plt.tight_layout()
plt.show()

```
